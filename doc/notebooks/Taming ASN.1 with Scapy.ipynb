## From ideas to binary

__Abstract Synax Notation One (ITU-T X.680)__

The ASN.1 notation defines a syntax for information data. It describes a number of simple
data types and specifies a notation for referencing these types and for specifying values of these types.

```
max-extensions INTEGER ::= 256
ExtensionSet ::= SET SIZE (1..max-extensions) OF Extension
BasicConstraints ::= SEQUENCE {cA BOOLEAN DEFAULT FALSE, pathLenConstraint INTEGER (0..MAX) OPTIONAL }
```

__Distinguished Encoding Rules (ITU-T X.690)__

The encoding of a data value shall consist of three components which shall appear in the following order:
- identifier octets;
- length octets;
- content octets.

For example, `my-ext-num INTEGER ::= 15` is encoded as `\x02 \x01 \x0f` with identifier=2 (integer) length=1 and content=15.

With Scapy, this example be encoded using:

````python
from scapy.all import *
my_ext_num = ASN1_INTEGER(15)
print(my_ext_num.val)
print(my_ext_num.enc(ASN1_Codecs.BER))
````

Similarly, a boolean can be encoded with:

````python
test_true = ASN1_BOOLEAN(True)
print(test_true.val)
print(test_true.enc(ASN1_Codecs.BER))
````

This binary values can be decoded with:

````python
ber_string = b'\x02\x01\x0f'
asn1_obj, remain = BERcodec_Object.dec(ber_string)
asn1_obj.show()
print(remain)
````

Several ASN.1 dialects coexist:
- Basic Encoding Rules
    - Distinguished Encoding Rules (a subset of BER) definite lengths only, unequivocal encoding for lengths, strings and sets
    - Canonical Encoding Rules (another subset of BER)
- XML Encoding Rules (encoding as XML strings)
- PER, OER, GSER...

## The right packet for the job

Let's say we need a packet for a BER-encoded simple extension.

````python
class ExtensionPkt(Packet):
    fields_desc = [ByteEnumField("type", 3, [0, 1, 2, 3]),
                   FieldLenField("len", None, length_of="extension_num"),
                   StrLenField("extension_num", "15", length_from=lambda pkt: pkt.len)]
````

But what about...

- a direct integer (i.e. not string) decoding?
- multibyte "type" or "length"?
- more complex ASN.1 data types?
- code conciseness?

Scapy provides seamless encoding through `ASN1F_*` fields: `ASN1F_INTEGER`, `ASN1F_BOOLEAN`, `ASN1F_STRING`...

````python
class ExtensionPkt(ASN1_Packet):
    ASN1_codec = ASN1_Codecs.BER
    ASN1_root = ASN1F_INTEGER("extension_num", 15)

p = ExtensionPkt()
print(raw(p))

p.extension_num = 256
print(raw(p))
````

Parsing is as intuitive as building:

````python
ExtensionPkt(b"\x02\x02\x01\x00").show()

ExtensionPkt(b"\x02\x01\x01\x00").show()

#ExtensionPkt(b"\x02\x03\x01\x00") # Note: not valid, it throws an error
````

Dubious packets can be forged, e.g. for validation purposes.

````python
class TrueOrFalsePkt(ASN1_Packet):
    ASN1_codec = ASN1_Codecs.BER
    ASN1_root = ASN1F_BOOLEAN("test", False)

p = TrueOrFalsePkt()
print(raw(p))

p.test = True
print(raw(p))

# Scapy lets you get naughty!
p.test = 0x0f
print(raw(p))
````

## Show me some CID


Object IDentifiers are usually given in the form of `X1.X2.[...].Xn`.

__DER encoding for OIDs (ITU-T X.690)__

- Let Y2 = 40 * X1 + X2 and Yi = Xi for i > 2
- With (ym , ym-1, ..., y0) as the base-128 writing of any Y, define the bytestream f(Y) = 1ym |1ym-1|...|1y1|0y0
- The binary value is the concatenatlon f(Y2)|f(Y3)|...|f(Yn)

For example, `countryName OBJECT IDENTIFIER ::= { 2 5 4 6 }` (the proper way to define 2.5.4.6) is encoded as `\x06 \x03 0101 0101 0000 0100 0000 0110` where id=6 (oid), length=3 and  content=f(85) f(4) f(6).

Scapy needs to know how to compute OIDs. You don't.

````python
class AttributeTypePkt(ASN1_Packet):
    ASN1_codec = ASN1_Codecs.BER
    ASN1_root = ASN1F_OID("at_type", "2.5.4.6")

p = AttributeTypePkt()
print(raw(p))

p.at_type = "2.5.128"
print(raw(p))

# Scapy actually knows several OIDs from its own MIB
print(conf.mib.countryName)

AttributeTypePkt(b"\x06\x03\x55\x04\x06").at_type
````

## Serenade for Strings

The ASN.1 syntax covers __many__ string constraints.

- BIT STRING: 3
- OCTET STRING: 4
- CHARACTER STRING: 29
- UTF8String: 12
- NumericString: 18
- PrintableString: 19
- TeletexString aka T615tring: 20
- VideotexString: 21
- A5String: 22
- GraphicString: 25
- VisibleString aka IS0646tring: 26
- GeneralString: 27
- UniversalString: 28
- BMPString: 30

However, for now, Scapy treats every string type the same (except for bit strings).

````python
class OctetStringPkt(ASN1_Packet):
    ASN1_codec = ASN1_Codecs.BER
    ASN1_root = ASN1F_STRING("msg", "")

p = OctetStringPkt(msg="hello world")
print(raw(p))

class PrintableStringPkt(ASN1_Packet):
    ASN1_codec = ASN1_Codecs.BER
    ASN1_root = ASN1F_PRINTABLE_STRING("msg", "")

# non-printable strings will not raise any warning
PrintableStringPkt("\x13\x06&@%*$#")
````

__DER encoding for bit strings (ITU-T X.690)__

- Let N be the number of excess bits used for byte-aligning the bit string (if it is empty, N shall be zero).
- The value always begins with N encoded as a binary integer.
- The subsequent bytes contain the byte-aligned bitstring.

For example, webserver-key-usage-f1ags BIT STRING 101 is encoded as `\x03 \x02 \x05 1010 0000` where id=3 (bit strinng), length=2, and content=`5 excess bits byte-aligned 101`

Because bit strings are frequently used for encoding data already byte-aligned, Scapy expects the field default to be a byte string.

````python
class BitStringPkt(ASN1_Packet):
    ASN1_codec = ASN1_Codecs.BER
    ASN1_root = ASN1F_BIT_STRING("msg", "hello")

p = BitStringPkt()
print(raw(p))

# this is the actual encoded bit string
print(p.msg.val)

# the byte string equivalent is stored in ‘val_readab1e’
print(p.msg.val_readable)
````

You may still provide the default as a bit string.

````python
class BitStringPkt2(ASN1_Packet):
    ASN1_codec = ASN1_Codecs.BER
    ASN1_root = ASN1F_BIT_STRING("msg", "101", default_readable=False)

p = BitStringPkt2()
print(raw(p))
print(p.msg.val)

# /!\ the ASN.1 class has the opposite default behaviour

p.msg = ASN1_BIT_STRING("hello", readable=True)
print(raw(p))
````
